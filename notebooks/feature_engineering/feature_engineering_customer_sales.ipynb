```python
#import libraries
import pandas as pd
import numpy as np

#load cleaned data
df = pd.read_csv("customers_cleaned.csv")

#age group feature
df["Age_Group"] = pd.cut(
df["Age"], bins=[0,25,35,45,55,65,100],
labels=[
"18-25",
"26-35",
"36-45",
"46-55",
"56-65",
"65+"    ])

#subscriber flag
df["Subscriber_Flag"] = np.where(df["Subscription Status"] == "Yes",1,0)

#customer value score
df["Customer_Value_Score"] = (df["Purchase Amount"]*(df["Subscriber_Flag"] + 1))

#purchase level
df["Purchase_Level"] = pd.qcut(df["Purchase Amount"],q=4,labels=[
"Low",
"Medium",
"High",
"Premium"])

#high value customer flag
df["High_Value_Customer"] = np.where (df["Purchase Amount"]>df["Purchase Amount"].quantile(0.75),1,0)

#profit estimation
df["Estimated_Profit"] = (df["Purchase Amount"] * 0.35)

#revenue contribution in %
total_revenue = df["Purchase Amount"].sum()

df["Revenue_Contribution_%"] = (df["Purchase Amount"] / total_revenue) * 100

# ==========================================
# SAVE FEATURED DATA
# ==========================================

df.to_csv(customers_featured.csv",index=False)
```
